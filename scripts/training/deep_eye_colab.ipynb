### 5. Configure Training Parameters

Sesuaikan parameter di bawah sebelum training.

| Parameter | T4 (15GB) | A100 (40GB) |
|-----------|-----------|-------------|
| batch-size | 2 | 4 |
| grad-accum | 4 | 2 |
| max-length | 2048 | 4096 |
| lora-r | 16 | 32 |
| epochs | 3 | 5 |

````python
# ========== TRAINING CONFIGURATION ==========
MODEL_NAME = "unsloth/qwen2.5-coder-7b-bnb-4bit"
DATASET_PATH = "data/pentest_dataset.jsonl"
OUTPUT_DIR = "models/deep-eye-hacker-7b"

# T4-friendly defaults (2GB VRAM overhead after model load)
EPOCHS = 3
BATCH_SIZE = 2
GRAD_ACCUM = 4
LEARNING_RATE = 2e-4
LORA_R = 16
LORA_ALPHA = 32
MAX_LENGTH = 2048
EXPORT_GGUF = True
# ============================================

print("Training configuration:")
print(f"  Model: {MODEL_NAME}")
print(f"  Epochs: {EPOCHS}, Batch: {BATCH_SIZE}, Grad Accum: {GRAD_ACCUM}")
print(f"  LoRA R={LORA_R}, Alpha={LORA_ALPHA}, LR={LEARNING_RATE}")
print(f"  Max Length: {MAX_LENGTH}")
print(f"  Export GGUF: {EXPORT_GGUF}")
````

### 6. Run Training

Memulai fine-tuning dengan Unsloth.  
**Estimasi waktu (T4, 7B model, 50-200 samples):** 15-45 menit per epoch.

````python
# ============================================================
# MAIN TRAINING LOOP
# This cell runs the actual fine-tuning
# ============================================================

import sys, os, json
from pathlib import Path

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

print("[*] Loading Unsloth + model...")

from unsloth import FastLanguageModel, is_bfloat16_supported
from unsloth import unsloth_train
from datasets import Dataset
from transformers import TrainingArguments
from trl import SFTTrainer

# Load 4-bit model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_NAME,
    max_seq_length=MAX_LENGTH,
    dtype=None,
    load_in_4bit=True,
)

# Add LoRA adapters
model = FastLanguageModel.get_peft_model(
    model,
    r=LORA_R,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha=LORA_ALPHA,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=42,
)

print(f"[*] Trainable parameters: {model.num_parameters(only_trainable=True):,}")

# Load dataset
def format_prompt(example: dict) -> str:
    return f"### Instruction:\n{example['instruction']}\n\n### Response:\n{example['output']}"

raw_data = []
with open(DATASET_PATH) as f:
    for line in f:
        item = json.loads(line)
        raw_data.append({"text": format_prompt(item)})

dataset = Dataset.from_list(raw_data)
print(f"[*] Dataset size: {len(dataset)} examples")

# Training arguments
training_args = TrainingArguments(
    per_device_train_batch_size=BATCH_SIZE,
    gradient_accumulation_steps=GRAD_ACCUM,
    warmup_ratio=0.1,
    num_train_epochs=EPOCHS,
    learning_rate=LEARNING_RATE,
    fp16=not is_bfloat16_supported(),
    bf16=is_bfloat16_supported(),
    logging_steps=10,
    optim="adamw_8bit",
    weight_decay=0.01,
    lr_scheduler_type="cosine",
    seed=42,
    output_dir=OUTPUT_DIR,
    report_to="none",
    ddp_find_unused_parameters=False,
)

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=MAX_LENGTH,
    dataset_num_proc=2,
    packing=True,
    args=training_args,
)

print("[*] Starting training...")
print(f"    Output dir: {OUTPUT_DIR}")
trainer.train()

# Save LoRA adapter
model.save_pretrained(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)
print(f"[+] LoRA adapter saved to: {OUTPUT_DIR}")
````

### 7. Export to GGUF (OLLAMA-compatible)

Merge LoRA weights into base model dan konversi ke GGUF format  
untuk digunakan dengan OLLAMA.

````python
# ============================================================
# MERGE + EXPORT to GGUF
# ============================================================

if EXPORT_GGUF:
    print("[*] Merging LoRA weights into base model...")

    import gc
    gc.collect()
    torch.cuda.empty_cache()
    merged_model = model.merge_and_unload()
    print("[+] Model merged successfully")

    # Save merged model in HF format
    MERGED_OUTPUT = OUTPUT_DIR + "-merged"
    merged_model.save_pretrained(MERGED_OUTPUT, safe_serialization=True)
    tokenizer.save_pretrained(MERGED_OUTPUT)
    print(f"[+] Merged model saved to: {MERGED_OUTPUT}")

    # Convert to GGUF
    print("[*] Converting to GGUF format (q8_0)...")
    GGUFPATH = f"{MERGED_OUTPUT}/deep-eye-hacker.gguf"
    import subprocess

    convert_cmd = [
        sys.executable, "-m", "llama_cpp.convert_hf_to_gguf",
        MERGED_OUTPUT,
        "--outfile", GGUFPATH,
        "--outtype", "q8_0",
    ]
    result = subprocess.run(convert_cmd, capture_output=True, text=True, timeout=600)

    if result.returncode == 0:
        gguf_size = os.path.getsize(GGUFPATH) / (1024**3)
        print(f"[+] GGUF model: {GGUFPATH} ({gguf_size:.2f} GB)")

        # Create OLLAMA Modelfile
        modelfile_lines = [
            f"FROM {GGUFPATH}",
            'TEMPLATE """### Instruction:\\n{{.Prompt}}\\n\\n### Response:\\n"""',
            "PARAMETER temperature 0.7",
            "PARAMETER top_p 0.95",
            'PARAMETER stop "### Instruction"',
            'SYSTEM "You are Deep Eye Hacker, an AI assistant specialized in penetration testing.\\nProvide accurate, technical responses for security testing, vulnerability detection, and exploit development."',
        ]
        modelfile = "\n".join(modelfile_lines)
        with open(f"{MERGED_OUTPUT}/Modelfile", "w") as f:
            f.write(modelfile)
        print(f"[+] OLLAMA Modelfile created")
    else:
        print(f"[!] GGUF conversion failed:")
        print(result.stderr[:500])

print("\n[DONE] Training + export complete!")
````

## Troubleshooting

| Problem | Solution |
|---------|----------|
| CUDA out of memory | Kurangi `BATCH_SIZE` ke 1, turunkan `MAX_LENGTH` ke 1024 |
| bitsandbytes error | Jalankan `!pip install --upgrade bitsandbytes` |
| llama.cpp import error | GGUF export gagal, pakai fallback HF format |
| Training terlalu lama | Kurangi `EPOCHS` ke 1 untuk quick test |
